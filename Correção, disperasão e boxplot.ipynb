```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

csv = pd.read_csv('Pokemon.csv')
df = pd.DataFrame(csv)
```

```python
pd.scatter_matrix(df, figsize = (25,25),diagonal = 'kde')
plt.show()
```

```python
#Verificando a correlação

df.corr(method='pearson')
```

```python
# Verificando ouliers com boxplot

box = df["HP"].plot(kind="box", grid=True)
print(df['HP'].quantile(0.97))
print(df['HP'].median())
plt.show()
```

```python
# calculando o valor dos 'bigodes'
# dado que respresantam 2 devios padrões

df['HP'].mean()+(df['HP'].std()*2)
```
