```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
import pandas as pd
```

```python
csv = pd.read_csv("Pokemon.csv")
df = pd.DataFrame(csv)
```

```python
df.head()
```

```python
# Verificando a distribuição do HP

#ordenando os valores
hp = df['HP'].sort_values()

#coletando a média e o desvio padrão
mu = hp.mean()
sigma = hp.std()

plt.plot(hp,mlab.normpdf(hp, mu, sigma))

plt.show()
```

```python
# Verificando a distribuição do Attack

#ordenando os valores
at = df['Attack'].sort_values()

#coletando a média e o desvio padrão
mu = at.mean()
sigma = at.std()

plt.figure(figsize=(15,15))
plt.plot(at, mlab.normpdf(at, mu, sigma), label="Desvio: {}".format(sigma))
plt.axvline(at.mean(), color='r', label="Média: {}".format(mu))
plt.legend(shadow=True)

plt.show()
```

```python
# Comparando distribuições

plt.figure(figsize=(15,15)) 


for col in df:
    if col != "Generation":
        if col != "#":
            if col != "Legendary":
                if col != "Total":
                    try:
                        s_col = df[col].sort_values()

                        mu = s_col.mean()
                        sigma = s_col.std()

                        plt.plot(s_col, mlab.normpdf(s_col, mu, sigma),label="{}, Média: {}".format(col,mu))

                        plt.legend(shadow=True)
                    except:
                        pass

plt.show()
```
